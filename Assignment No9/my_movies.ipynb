```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt

# Load the datasets
book_data = pd.read_csv('book.csv')
movies_data = pd.read_csv('my_movies.csv')

# Convert the datasets to a one-hot encoded format
book_data_encoded = pd.get_dummies(book_data)
movies_data_encoded = pd.get_dummies(movies_data)

# Apply Apriori algorithm and generate association rules
def generate_rules(data, min_support, min_confidence, min_length):
    frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
    rules = rules[rules['antecedents'].apply(lambda x: len(x) >= min_length)]
    return rules

# Visualize the obtained rules using different plots
def visualize_rules(rules):
    if 'support' in rules.columns and 'confidence' in rules.columns:
        plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
        plt.xlabel('Support')
        plt.ylabel('Confidence')
        plt.title('Support vs Confidence')
        plt.show()
    else:
        print("The rules DataFrame does not contain 'support' and 'confidence' columns.")

# Example usage
support_values = [0.1, 0.2, 0.3]  # Example support values to try
confidence_values = [0.5, 0.6, 0.7]  # Example confidence values to try
min_length_values = [2, 3]  # Example minimum length values to try

for min_support in support_values:
    for min_confidence in confidence_values:
        for min_length in min_length_values:
            rules = generate_rules(book_data_encoded, min_support, min_confidence, min_length)
            print(f"Number of rules for support={min_support}, confidence={min_confidence}, min_length={min_length}: {len(rules)}")
            visualize_rules(rules)
```
