```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bank-full.csv", delimiter=";")
```

```python
print(df.head())  # Display first few rows
print(df.info())  # Summary of DataFrame
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='y')
plt.title("Term Deposit Subscription")
plt.xlabel("Subscribed to Term Deposit?")
plt.ylabel("Count")
plt.show()
```

```python
numerical_vars = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
plt.figure(figsize=(12, 8))
for i, var in enumerate(numerical_vars, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data=df, x=var, bins=20, kde=True)
    plt.title(f"Distribution of {var.capitalize()}")
    plt.xlabel(var.capitalize())
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
categorical_vars = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
plt.figure(figsize=(15, 10))
for i, var in enumerate(categorical_vars, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=var, hue='y', palette='muted')
    plt.title(f"{var.capitalize()} vs Term Deposit Subscription")
    plt.xlabel(var.capitalize())
    plt.ylabel("Count")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
